```julia
using Plots
pyplot()
```

```julia
function lorenz(x, y, z, sigma=10, rho=28, beta=2.667)
    
    xdot = sigma*(y - x)
    ydot = x*(rho - z) - y
    zdot = x*y - beta*z
    
    [xdot, ydot, zdot]
    end;
```

```julia
function make_attractor(nt, dt, x1, y1, z1)
    r = [zeros(nt), zeros(nt), zeros(nt)]
    r[1][1], r[2][1], r[3][1] = x1, y1, z1
    for i in 1:nt-1
        xdot, ydot, zdot = lorenz(r[1][i], r[2][i], r[3][i])
        r[1][i + 1] = r[1][i] + (xdot * dt)
        r[2][i + 1] = r[2][i] + (ydot * dt)
        r[3][i + 1] = r[3][i] + (zdot * dt)
    end
    r
end;
```

```julia
dt = 0.01
nt = 10000

x1, y1, z1 = 1.0, 0.0, 0.0
r = make_attractor(nt, dt, x1, y1, z1)
r2 = make_attractor(nt, dt, x1, y1, z1+0.1)
```

```julia
cmap = range(0, stop=1, length=10000)

p = plot(plot3d(r[1], r[2], r[3], line_z=cmap, linetype=:path3d, legend=false, title="Lorenz Attractor", linecolor=:plasma), 
    plot3d(r2[1], r2[2], r2[3], line_z=cmap, linetype=:path3d, legend=false, title="Lorenz Attractor 2", linecolor=:blues), 
    layout=2)
#p = plot3d(r[1], r[2], r[3], line_z=cmap, linetype=:path3d, legend=false, title="Lorenz Attractor")
```
